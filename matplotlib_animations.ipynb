```python
from matplotlib.animation import FuncAnimation
from matplotlib import pyplot
import numpy

from mpl_toolkits.mplot3d.art3d import Line3D
from matplotlib.lines import Line2D

from IPython import display
```

```python
x = numpy.linspace(0, 2*numpy.pi, 100)

def animate(frame: int, plot: Line2D):
    
    # Update plot
    line = plot[0]
    y = numpy.sin(x + 2 * numpy.pi * frame/100)
    line.set_data((x,y))
```

```python
fig = pyplot.figure()
ax = pyplot.axes()
plt = pyplot.plot([])
pyplot.xlim(0, 2 * numpy.pi)
pyplot.ylim(-2,2)
pyplot.close()
```

```python
''' frames: Number of Frames to be shown'''
'''interval: Interval between the frames in milli Sec'''
animation = FuncAnimation(fig = fig, func = animate, frames = 100 , interval = 20, fargs = [plt] ) 

video = animation.to_html5_video()
html = display.HTML(video)
display.display(html)
pyplot.close()
```

```python
z = numpy.linspace(0, 15, 1000)

def animate(frame: int, plot1: Line3D, plot2: Line3D):
    
    # Update plot
    line1 = plot1[0]
    x = numpy.sin(10 * z + 2 * numpy.pi * frame/100)
    y = numpy.cos(10 * z + 2 * numpy.pi * frame/100)
    line1.set_data_3d(x, y, z)
    line1.set_color("red")
    
    line2 = plot2[0]
    x = numpy.sin(10 * z - 2 * numpy.pi * frame/100)
    y = numpy.cos(10 * z - 2 * numpy.pi * frame/100)
    line2.set_data_3d(x, y, z)
    line2.set_color("blue")
```

```python
fig = pyplot.figure()
ax = pyplot.axes(projection ='3d')
plt1 = pyplot.plot([],[])
plt2 = pyplot.plot([],[])
pyplot.xlim(0, 2 * numpy.pi)
pyplot.ylim(-2,2)
pyplot.close()
```

```python
''' frames: Number of Frames to be shown'''
'''interval: Interval between the frames in milli Sec'''
animation = FuncAnimation(fig = fig, func = animate, frames = 100 , interval = 20, fargs = [plt1,plt2] ) 

video = animation.to_html5_video()
html = display.HTML(video)
display.display(html)
pyplot.close()
```
